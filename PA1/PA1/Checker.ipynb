```python
import numpy as np
from scipy import stats
a = np.random.randint(3,size=(5,5))
print(a,"\n\n")
c = stats.mode(a,axis=1,keepdims=True)
print(c)
print(c[0].T[0])


b = np.argsort(a)
print(b)
print(b<3)
```
