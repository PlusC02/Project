```python
def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)


df["Email"] = df["Email"].apply(lambda x : remove_html(x))
```

```python
#Lowercasting
df["Email"] = df["Email"].str.lower()
```

```python
df.head()
```

```python
#Remove urls
import re
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)


df["Email"] = df["Email"].apply(lambda text: remove_urls(text))
df.head()
```

```python
#Punctual removal
import string
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["Email"] = df["Email"].apply(lambda x : remove_punctuation(x))
print(df["Email"][1])
```

```python
import nltk
from nltk.corpus import stopwords

#nltk.download('stopwords')
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])


df["Email"] = df["Email"].apply(lambda x : remove_stopwords(x))
print("Execuated")
```

```python
#Lemmatiaztion
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

df["Email"] = df["Email"].apply(lambda text: lemmatize_words(text))
```
